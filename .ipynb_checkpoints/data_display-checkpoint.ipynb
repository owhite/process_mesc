```python
import urllib.request
import json
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import hello

def get_web_page(req):
    try: urllib.request.urlopen(req)
    except urllib.error.URLError as e:
        print(e.reason) 
    with urllib.request.urlopen(req) as response:
        the_page = response.read().decode('utf-8')
        response.close()
        return(the_page)
        
def make_frame(page): 
    lines = page.splitlines();

    names = []
    sets = []
    i = 0

    for key in json.loads(lines[0]).keys():
        names.append(key)
        sets.append([])
        i = i + 1

    count = 0
    x = []
    for line in lines:
        js = json.loads(line)
        x.append(count)
        count = count + 1
        i = 0
        for n in names:
            sets[i].append(js[n])
            i = i + 1

    my_dict = {}
    i = 0
    for key in json.loads(lines[0]).keys():
        thing = {names[i]: sets[i]}
        my_dict.update(thing)
        i = i + 1
        
    return(my_dict)

# URL = input()
URL = 'https://raw.githubusercontent.com/owhite/ebike_data/main/first_set'
page = get_web_page(URL)
data = make_frame(page)

df = pd.DataFrame(data)
hello.world()

df['phaseA'] = np.sqrt( (df['idq_d'] * df['idq_d']) + (df['idq_q'] * df['idq_q']) )

plt.plot(df['ehz'])
plt.plot(df['phaseA'])

plt.show()    
```
